# Read the image and creating the pencil sketch

```python
import cv2
import matplotlib.pyplot as plt

def pencil_sketch(image_path):
    # Read the image
    image = cv2.imread(image_path)
    
    # Convert image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Invert the grayscale image
    inverted_gray_image = 255 - gray_image
    
    # Blur the inverted grayscale image
    blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)
    
    # Invert the blurred image
    inverted_blurred_image = 255 - blurred_image
    
    # Create the pencil sketch
    pencil_sketch_image = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)
    
    return gray_image, inverted_gray_image, blurred_image, inverted_blurred_image, pencil_sketch_image

# Path to input image
input_image_path = "C:\\Users\\kmpvi\\Pictures\\morish.jpg"

# Generate intermediate and final images
gray, inv_gray, blur, inv_blur, sketch = pencil_sketch(input_image_path)


plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(cv2.imread(input_image_path), cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()


```

# Convert the image into greyscale

```python
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

```

# Invert the greyscale image (Negative image)

```python
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(inv_gray, cv2.COLOR_BGR2RGB))
plt.title('Inverted Grayscale Image')
plt.axis('off')
plt.show()
```

# Blurred image

```python
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image')
plt.axis('off')
plt.show()

```

# Inverted blurry image

```python
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(inv_blur, cv2.COLOR_BGR2RGB))
plt.title('Inverted Blurred Image')
plt.axis('off')
plt.show()

```

# Pencil sketch image

```python
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
plt.title('Pencil Sketch')
plt.axis('off')
plt.show()
```

# Comparing original and pencil sketch image

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cv2.imread(input_image_path), cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
plt.title('Pencil Sketch')
plt.axis('off')

plt.show()
```
